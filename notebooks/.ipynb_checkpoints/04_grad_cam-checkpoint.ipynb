```python
# ==============================================================
# Occlusion Sensitivity explainability (works for your nested model)
# ==============================================================
import numpy as np
import matplotlib.pyplot as plt
import cv2
import os
import tensorflow as tf
from tensorflow.keras.models import load_model
from tqdm import tqdm

# 1) Load model (use your path)
model = load_model(r"C:\Users\vgang\medical-image-classification\models\efficientnet_b0_finetuned.keras")
print("Loaded model. Top layers:", [l.name for l in model.layers])

# 2) Pick test image
test_dir = r"C:\Users\vgang\medical-image-classification\data\chest_xray\test\PNEUMONIA"
img_path = os.path.join(test_dir, os.listdir(test_dir)[0])
print("Using:", img_path)

# 3) Load & preprocess
orig = tf.keras.utils.load_img(img_path, target_size=(224,224))
orig_arr = tf.keras.utils.img_to_array(orig) / 255.0
inp = np.expand_dims(orig_arr, axis=0).astype(np.float32)

# baseline prediction (model outputs probability ~ [0,1] for PNEUMONIA)
base_pred = float(model.predict(inp)[0][0])
print(f"Baseline prediction (Pneumonia score): {base_pred:.4f}")

# 4) Occlusion parameters (tune these for speed/quality)
patch_size = 32      # size of the square occluder
stride = 16          # how far to move the occluder each step
img_h, img_w = 224, 224

# compute number of positions
h_steps = (img_h - patch_size) // stride + 1
w_steps = (img_w - patch_size) // stride + 1

heatmap = np.zeros((h_steps, w_steps), dtype=np.float32)

# 5) For speed, we create a batch per row (optional). We'll loop reasonably.
for i in tqdm(range(h_steps), desc="Rows"):
    batch_imgs = []
    positions = []
    for j in range(w_steps):
        y = i * stride
        x = j * stride
        occluded = orig_arr.copy()
        # occlude region with mean color (or 0)
        occluded[y:y+patch_size, x:x+patch_size, :] = np.mean(orig_arr, axis=(0,1))
        batch_imgs.append(occluded)
        positions.append((i,j))
    batch = np.stack(batch_imgs, axis=0).astype(np.float32)
    preds = model.predict(batch, verbose=0).squeeze()  # shape (batch,)
    # compute effect = base_pred - occluded_pred  (positive means occlusion lowered score)
    effects = base_pred - preds
    for (i_pos,j_pos), eff in zip(positions, effects):
        heatmap[i_pos, j_pos] = eff

# 6) Upsample heatmap to image size
heatmap_full = cv2.resize(heatmap, (img_w, img_h), interpolation=cv2.INTER_CUBIC)
# normalize between 0 and 1
heatmap_full = np.maximum(heatmap_full, 0)
if heatmap_full.max() > 0:
    heatmap_full = heatmap_full / heatmap_full.max()
else:
    heatmap_full = heatmap_full

# 7) Create colored overlay
heatmap_uint8 = np.uint8(255 * heatmap_full)
heatmap_color = cv2.applyColorMap(heatmap_uint8, cv2.COLORMAP_JET)  # BGR
overlay = cv2.addWeighted(cv2.cvtColor((orig_arr*255).astype(np.uint8), cv2.COLOR_RGB2BGR), 
                          0.7, heatmap_color, 0.3, 0)
overlay_rgb = cv2.cvtColor(overlay, cv2.COLOR_BGR2RGB)

# 8) Display original, heatmap, overlay
plt.figure(figsize=(14,5))

plt.subplot(1,3,1)
plt.imshow(orig)
plt.title("Original X-ray")
plt.axis('off')

plt.subplot(1,3,2)
plt.imshow(heatmap_full, cmap='jet')
plt.title("Occlusion heatmap (normalized)")
plt.axis('off')

plt.subplot(1,3,3)
plt.imshow(overlay_rgb)
plt.title(f"Overlay (baseline score={base_pred:.3f})")
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
# ==============================================================
# Test multiple X-rays (both NORMAL and PNEUMONIA)
# ==============================================================

# pick some images from both classes
normal_dir = r"C:\Users\vgang\medical-image-classification\data\chest_xray\test\NORMAL"
pneu_dir   = r"C:\Users\vgang\medical-image-classification\data\chest_xray\test\PNEUMONIA"

# you can adjust how many examples to test
normal_imgs = [os.path.join(normal_dir, f) for f in os.listdir(normal_dir)[:2]]
pneu_imgs   = [os.path.join(pneu_dir, f) for f in os.listdir(pneu_dir)[:2]]

test_imgs = normal_imgs + pneu_imgs

# function to compute occlusion map for a single image (reuse from before)
def occlusion_map(img_path, model, patch_size=32, stride=16):
    orig = tf.keras.utils.load_img(img_path, target_size=(224,224))
    arr = tf.keras.utils.img_to_array(orig) / 255.0
    inp = np.expand_dims(arr, 0).astype(np.float32)
    base_pred = float(model.predict(inp, verbose=0)[0][0])

    img_h, img_w = 224, 224
    h_steps = (img_h - patch_size) // stride + 1
    w_steps = (img_w - patch_size) // stride + 1
    heatmap = np.zeros((h_steps, w_steps), dtype=np.float32)

    for i in range(h_steps):
        row_imgs = []
        positions = []
        for j in range(w_steps):
            y, x = i * stride, j * stride
            occluded = arr.copy()
            occluded[y:y+patch_size, x:x+patch_size, :] = np.mean(arr, axis=(0,1))
            row_imgs.append(occluded)
            positions.append((i,j))
        batch = np.stack(row_imgs, axis=0)
        preds = model.predict(batch, verbose=0).squeeze()
        effects = base_pred - preds
        for (i_pos,j_pos), eff in zip(positions, effects):
            heatmap[i_pos, j_pos] = eff

    heatmap = cv2.resize(heatmap, (img_w, img_h))
    heatmap = np.maximum(heatmap, 0)
    if heatmap.max() > 0:
        heatmap /= heatmap.max()

    heatmap_uint8 = np.uint8(255 * heatmap)
    heatmap_color = cv2.applyColorMap(heatmap_uint8, cv2.COLORMAP_JET)
    overlay = cv2.addWeighted(
        cv2.cvtColor((arr*255).astype(np.uint8), cv2.COLOR_RGB2BGR), 0.7,
        heatmap_color, 0.3, 0
    )
    overlay = cv2.cvtColor(overlay, cv2.COLOR_BGR2RGB)
    return orig, overlay, base_pred

# ==============================================================
# Show results for several images
# ==============================================================

for path in test_imgs:
    orig, overlay, score = occlusion_map(path, model)
    label = "PNEUMONIA" if "PNEUMONIA" in path.upper() else "NORMAL"
    plt.figure(figsize=(10,5))
    plt.subplot(1,2,1)
    plt.imshow(orig)
    plt.title(f"Original ({label})")
    plt.axis("off")
    plt.subplot(1,2,2)
    plt.imshow(overlay)
    plt.title(f"Occlusion Overlay\nPredicted score: {score:.3f}")
    plt.axis("off")
    plt.show()
```
