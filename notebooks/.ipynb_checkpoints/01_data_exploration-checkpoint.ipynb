```python
import os
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
base_dir = r"C:\Users\vgang\medical-image-classification\data\chest_xray"
train_dir = os.path.join(base_dir, "train")
test_dir = os.path.join(base_dir, "test")
val_dir = os.path.join(base_dir, "val")

for folder in [train_dir, test_dir, val_dir]:
    print(folder, "->", os.listdir(folder))
```

```python
for split in ['train', 'test', 'val']:
    split_dir = os.path.join(base_dir, split)
    for cls in os.listdir(split_dir):
        count = len(os.listdir(os.path.join(split_dir, cls)))
        print(f"{split}/{cls}: {count} images")
```

```python
import random
from tensorflow.keras.utils import load_img, img_to_array

def show_samples(split='train', cls='PNEUMONIA', num=5):
    folder = os.path.join(base_dir, split, cls)
    files = random.sample(os.listdir(folder), num)
    plt.figure(figsize=(15,5))
    for i, fname in enumerate(files):
        img_path = os.path.join(folder, fname)
        img = load_img(img_path)
        plt.subplot(1, num, i+1)
        plt.imshow(img)
        plt.axis('off')
    plt.suptitle(f"Examples from {split}/{cls}", fontsize=16)
    plt.show()

show_samples('train', 'NORMAL', 5)
show_samples('train', 'PNEUMONIA', 5)
```

```python
img_size = (224, 224)
batch_size = 32

train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=15,
    width_shift_range=0.1,
    height_shift_range=0.1,
    zoom_range=0.1,
    horizontal_flip=True
)

val_datagen = ImageDataGenerator(rescale=1./255)

train_gen = train_datagen.flow_from_directory(
    train_dir,
    target_size=img_size,
    batch_size=batch_size,
    class_mode='binary'
)

val_gen = val_datagen.flow_from_directory(
    val_dir,
    target_size=img_size,
    batch_size=batch_size,
    class_mode='binary'
)

test_gen = val_datagen.flow_from_directory(
    test_dir,
    target_size=img_size,
    batch_size=batch_size,
    class_mode='binary',
    shuffle=False
)
```
